# KNN from Scratch

```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
```

## Data Input

```python
data_csv = pd.read_table("/home/user/Documents/ml-100k/u.data",header=None,names=['user id','movie id','rating','timestamp'])
```

```python
l_user = (data_csv['user id'].unique())
l_movie = (data_csv['movie id'].unique())
```

```python
sample = np.zeros((len(l_user),len(l_movie))) 
```

```python
for line in data_csv.itertuples():
    sample[line[1]-1, line[2]-1] = line[3]
```

## Similarity Calculation

```python
sim = cosine_similarity(sample)
```

## Mean Calculation

```python
x=sim.argsort(axis=1)
sample1=sample.copy()
z=np.count_nonzero([sample1.T],axis=1)
mean = sample1.sum(axis=1)/z
mean_mat1 = mean.T
```

## Calculating difference between Rating and avg. rating

```python
rat_dif = np.zeros((943,1682))
for i in range(sample.shape[0]):
    for j in range(sample.shape[1]):
        if sample[i,j]==0:
            rat_dif[i,j] = 0
        else:
            rat_dif[i,j] = sample[i,j]  - mean_mat1[i]
```

## Calculation of Top Similarities

```python
top_10_similar_users = x[:, -7 : ]
```

## Prediction Algorithm

```python
# using mean for the prediction
collective_sum = np.zeros((943, 1682))
pred = np.zeros((943,1682))
count=0
for i in (top_10_similar_users):
    numerator = np.zeros((1,1682))
    denominator = 0
    for j in i[0:-1]:
        denominator += sim[i[-1],[j]]
        numerator = numerator + sim[i[-1],[j]] * (rat_dif[j, :])
    collective_sum[count,:] = numerator / denominator 
    pred[count,:] = collective_sum[count,:] + mean_mat1[count]
    count+=1
    
print(pd.DataFrame(pred))   
```

```python
# without using mean for the prediction
collective_sum = np.zeros((943, 1682))
pred = np.zeros((943,1682))
count=0
for i in (top_10_similar_users):
    numerator = np.zeros((1,1682))
    denominator = 0
    for j in i[0:-1]:
        denominator += sim[i[-1],[j]]
        numerator = numerator + sim[i[-1],[j]] * (sample[j, :])
    
    value = numerator / denominator 
    pred[count,:] = value 
    count+=1
print(pd.DataFrame(pred))
```
